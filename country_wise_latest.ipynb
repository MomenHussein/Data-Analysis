```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("country_wise_latest.csv")
x=data['Recovered']
y=data['Deaths']
plt.rcParams["axes.prop_cycle"] = plt.cycler(
    color=["#4C2A85", "#BE96FF", "#957DAD", "#5E366E", "#A98CCC"])

fig1, ax1 = plt.subplots()
x= data["Deaths"]
y1=data["Recovered"]
y2=data["Confirmed"]
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='o')
ax1.set_title("Recovered/Deaths legend")
ax1.set_xlabel("Recovered")
ax1.set_ylabel("Deaths")
 
# Labels and Title
plt.xlabel('Deaths'), plt.ylabel('Recovered'), plt.title('Deaths/Recovered')
 
# Legend and Display
plt.legend()


fig2, ax2 = plt.subplots()
x= data["Deaths"]
y=data["Recovered"]
plt.plot(x, y, 'o')
ax2.set_title("Recovered/Deaths legend")
ax2.set_xlabel("Recovered")
ax2.set_ylabel("Deaths")
# plt.show()


# plt.show()

fig4, ax4 = plt.subplots()
ax4.plot(list(x), list(y))
ax4.set_title("Recovered/Deaths")
ax4.set_xlabel("Recovered")
ax4.set_ylabel("Deaths")
# plt.show()

fig5, ax5 = plt.subplots()
ax5.fill_between(x,y)
ax5.set_title("Recovered/Deaths sub")
ax5.set_xlabel("Recovered")
ax5.set_ylabel("Deaths")
# plt.show()

# Create a window and add charts
root = tk.Tk()
root.title("Dashboard")
root.state('zoomed')

side_frame = tk.Frame(root, bg="#4C2A85")
side_frame.pack(side="left", fill="y")

label = tk.Label(side_frame, text="Dashboard", bg="#4C2A85", fg="#FFF", font=25)
label.pack(pady=50, padx=20)

charts_frame = tk.Frame(root)
charts_frame.pack()

upper_frame = tk.Frame(charts_frame)
upper_frame.pack(fill="both", expand=True)

canvas1 = FigureCanvasTkAgg(fig1, upper_frame)
canvas1.draw()
canvas1.get_tk_widget().pack(side="left", fill="both", expand=True)

canvas2 = FigureCanvasTkAgg(fig2, upper_frame)
canvas2.draw()
canvas2.get_tk_widget().pack(side="left", fill="both", expand=True)

lower_frame = tk.Frame(charts_frame)
lower_frame.pack(fill="both", expand=True)

canvas4 = FigureCanvasTkAgg(fig4, lower_frame)
canvas4.draw()
canvas4.get_tk_widget().pack(side="left", fill="both", expand=True)

canvas5 = FigureCanvasTkAgg(fig5, lower_frame)
canvas5.draw()
canvas5.get_tk_widget().pack(side="left", fill="both", expand=True)

root.mainloop()
```

```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("country_wise_latest.csv")
x=data['Recovered']
y=data['Deaths']
plt.rcParams["axes.prop_cycle"] = plt.cycler(
    color=["#4C2A85", "#BE96FF", "#957DAD", "#5E366E", "#A98CCC"])

fig3, ax3 = plt.subplots()
country_deaths = data[['Country/Region', 'Recovered']].sort_values(by='Recovered', ascending=False)
plt.bar(country_deaths['Country/Region'].head(10), country_deaths['Recovered'].head(10))
plt.xticks(rotation=80)
ax3.set_title("Recovered/Deaths Bar")
ax3.set_xlabel("Recovered")
ax3.set_ylabel("Deaths")

fig6, ax6 = plt.subplots()
recovered = data["Recovered"].sum()
death = data["Deaths"].sum()
data = [recovered, death]
labels = ["Recovered cases", "Deaths"]
ax6.set_title("Recovered/Deaths Pie")
ax6.set_xlabel("Recovered")
ax6.set_ylabel("Deaths")

pie = plt.pie(data, labels=labels, autopct="%.2f%%")

root = tk.Tk()
root.title("Dashboard")
root.state('zoomed')

side_frame = tk.Frame(root, bg="#4C2A85")
side_frame.pack(side="left", fill="y")

label = tk.Label(side_frame, text="Dashboard", bg="#4C2A85", fg="#FFF", font=25)
label.pack(pady=50, padx=20)

charts_frame = tk.Frame(root)
charts_frame.pack()

upper_frame = tk.Frame(charts_frame)
upper_frame.pack(fill="both", expand=True)

canvas3 = FigureCanvasTkAgg(fig3, upper_frame)
canvas3.draw()
canvas3.get_tk_widget().pack(side="right", fill="both", expand=True)


canvas6 = FigureCanvasTkAgg(fig6, upper_frame)
canvas6.draw()
canvas6.get_tk_widget().pack(side="right", fill="both", expand=True)

lower_frame = tk.Frame(charts_frame)
lower_frame.pack(fill="both", expand=True)


root.mainloop()
```
